# Setup

### Imports<a class="anchor" id="H0_imports"></a>

```python
# Appending the package's location to path
import sys
sys.path.append('..')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Fimod import
from fimod import Stock, CallOption, PutOption
from fimod import plot_payoff
```

### Settings<a class="anchor" id="H0_settings"></a>

```python
# Matplotlib settings
%matplotlib inline
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (10.0, 8.0)
```

# Option Strategies<a class="anchor" id="H1"></a>

```python
kwargs = dict(
    invline_kwds={"alpha":0.5},
    portfline_kwds={"color":"k"},
    pricerange=np.arange(10,50)
)
```

## Bull & Bear Spreads Using Calls<a class="anchor" id="H11"></a>

```python
plot_payoff(
    CallOption(strike=20, premium=10, position="long"),
    CallOption(strike=30, premium=5, position="short"),
    title="Bull Spread with calls",
    **kwargs
)
plt.figure()
plot_payoff(
    CallOption(strike=20, premium=10, position="short"),
    CallOption(strike=30, premium=5, position="long"),
    title="Bear Spread with calls",
    **kwargs
)
```

## Bull & Bear Spreads Using Puts<a class="anchor" id="H12"></a>

```python
plot_payoff(
    PutOption(strike=20, premium=5, position="long"),
    PutOption(strike=30, premium=10, position="short"),
    title="Bull Spread with Puts",
    **kwargs
)

plt.figure()
plot_payoff(
    PutOption(strike=20, premium=5, position="short"),
    PutOption(strike=30, premium=10, position="long"),
    title="Bear Spread with Puts",
    **kwargs
)
```

## Straddle<a class="anchor" id="H13"></a>

```python
plot_payoff(
    CallOption(strike=30, premium=5, position="long"),
    PutOption(strike=30, premium=10, position="long"),
    title="Straddle",
    **kwargs
)
```

## Strangle<a class="anchor" id="H14"></a>

```python
plot_payoff(
    CallOption(strike=25, premium=5, position="long"),
    PutOption(strike=35, premium=10, position="long"),
    title="Strangle",
    **kwargs
)
```

## Strip & Strap<a class="anchor" id="H15"></a>

```python
plot_payoff(
    CallOption(strike=30, premium=5, position="long"),
    PutOption(strike=30, premium=5, position="long"),
    PutOption(strike=30, premium=5, position="long"),
    title="Strip",
    **kwargs
)

plt.figure()
plot_payoff(
    CallOption(strike=30, premium=5, position="long"),
    CallOption(strike=30, premium=5, position="long"),
    PutOption(strike=30, premium=5, position="long"),
    title="Strap",
    **kwargs
)
```

## Butterfly<a class="anchor" id="H16"></a>

```python
plot_payoff(
    CallOption(strike=25, premium=7, position="long"),
    CallOption(strike=35, premium=5, position="long"),
    CallOption(strike=30, premium=5, position="short"),
    CallOption(strike=30, premium=5, position="short"),
    title="Butterfly",
    **kwargs
)
```
